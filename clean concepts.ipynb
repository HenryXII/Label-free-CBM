```python
import json
# Function to read JSON file
def read_json_file(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

# Function to merge dictionaries
def merge_dicts(dict_list):
    merged_dict = {}
    for d in dict_list:
        for key, value in d.items():
            if key in merged_dict:
                merged_dict[key].extend(value)
            else:
                merged_dict[key] = value.copy()
    return merged_dict

dataset=""

# Input file paths
file_path1 = "data/concept_sets/gpt3_init/gpt3_{}_around.json".format(dataset)
file_path2 = "data/concept_sets/gpt3_init/gpt3_{}_important.json".format(dataset)
file_path3 = "data/concept_sets/gpt3_init/gpt3_{}_superclass.json".format(dataset)

# Read JSON files
data_list = [read_json_file(file_path) for file_path in [file_path1, file_path2, file_path3]]

# Merge the dictionaries
merged_data = merge_dicts(data_list)

# Remove duplicates from each list in the merged dictionary
for key in merged_data:
    merged_data[key] = list(set(merged_data[key]))

# Sort the lists for each key
for key in merged_data:
    merged_data[key].sort()

# Save the merged data to a new JSON file
output_file = 'data/concept_sets/{}_acc.json'.format(dataset)
with open(output_file, 'w') as outfile:
    json.dump(merged_data, outfile, indent=4)

print(f"Merged data has been saved to '{output_file}'")

# Display the merged data
print("\nMerged Data:")
print(json.dumps(merged_data, indent=4))
```

```python
import json

def read_json(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

def read_txt(file_path):
    with open(file_path, 'r') as file:
        return set(line.strip() for line in file)

def filter_json(json_data, word_list):
    return {key: [item for item in value if item in word_list] for key, value in json_data.items()}

def write_json(data, file_path):
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)

dataset="cifar10"
json_file = 'data/concept_sets/{}_acc.json'.format(dataset)
txt_file = 'data/concept_sets/{}_manual.txt'.format(dataset)

json_data = read_json(json_file)
word_list = read_txt(txt_file)

filtered_data = filter_json(json_data, word_list)

write_json(filtered_data, json_file)
```
